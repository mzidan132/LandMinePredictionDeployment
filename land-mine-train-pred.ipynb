```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

X = mine.iloc[:, :-1].values
y = mine.iloc[:, -1].values
```

```python
X_train = X[:225,:]
X_test = X[225:,:]
y_train = y[:225]
y_test = y[225:]
print(X_test, len(X_test))
print(y_test, len(y_test))
```

```python
scaler = StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn import svm
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

models = []
models.append(SVC(kernel='rbf', gamma=0.67, C=31))
models.append(LogisticRegression())
models.append(DecisionTreeClassifier())
models.append(RandomForestClassifier(n_estimators=10))
models.append(AdaBoostClassifier())
models.append(GaussianNB())
```

```python
from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

model_list = ['SVM','LogisticReg','Decision Tree','Random Forest','Adaboost','Naive Bayes']
acc_train_list = []
acc_test_list = []
cm_list = []

for model in models:
    model.fit(X_train, y_train)
    y_pred_train = model.predict(X_train)
    y_pred_test = model.predict(X_test)
    acc_train_list.append(metrics.accuracy_score(y_train, y_pred_train))
    acc_test_list.append(metrics.accuracy_score(y_test, y_pred_test))
    cm_list.append(confusion_matrix(y_test, y_pred_test))
```

```python
import pandas as pd
import math

# Convert accuracy values to integers (rounded up)
acc_train_list_int = [math.ceil(acc * 100) for acc in acc_train_list]
acc_test_list_int = [math.ceil(acc * 100) for acc in acc_test_list]

# Create DataFrame
result_df = pd.DataFrame({'Model': model_list, 'Train Accuracy': acc_train_list_int, 'Test Accuracy': acc_test_list_int})

# Display the accuracy DataFrame
print(result_df)
```

```python
from sklearn.neural_network import MLPClassifier
mlp_clf = MLPClassifier(random_state=55)
mlp_clf.fit(X_train, y_train)

# Adaboost with Decision Tree Classifier
ada_rf_clf = AdaBoostClassifier(base_estimator=DecisionTreeClassifier(random_state=55), n_estimators=50, random_state=55)
ada_rf_clf.fit(X_train, y_train)

# Make predictions

mlp_pred = mlp_clf.predict(X_test)
ada_rf_pred = ada_rf_clf.predict(X_test)
#
mlp_accuracy = accuracy_score(y_test, mlp_pred)
ada_rf_accuracy = accuracy_score(y_test, ada_rf_pred)

print(f"MLP Accuracy: {mlp_accuracy}")
print(f"Adaboost with Decision Tree Accuracy: {ada_rf_accuracy}")
```

```python
# Define a dictionary to map numerical categories to their names
category_mapping = {
    1: "Not-Mine",
    2: "Anti-tank",
    3: "Anti-personnel",
    4: "Booby-Trapped-Anti-personnel",
    5: "M14-Anti-personnel"
}

def make_prediction(model, scaler, voltage, height, soil):
    # Create a numpy array with the input features
    new_data = np.array([[voltage, height, soil]])

    # Scale the new data using the same scaler used for training
    scaled_data = scaler.transform(new_data)

    # Make predictions using the specified model
    prediction = model.predict(scaled_data)

    # Map numerical category to its name
    predicted_category = category_mapping[prediction[0]]

    return predicted_category

# Example usage with the first SVM model
svm_model = models[0]  # Assuming SVM is the first model

# Example input values
voltage_input = 0.32
height_input = 0.8181
soil_input = 2

# Make a prediction using the first SVM model
predicted_mine = make_prediction(svm_model, scaler, voltage_input, height_input, soil_input)

# Display the predicted mine category
print(f"The predicted mine category is: {predicted_mine}")
```
